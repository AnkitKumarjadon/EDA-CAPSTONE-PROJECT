# Handling null/missing values in dataset

```python
Null_Values_Find = pd.DataFrame(df.isna().sum(),columns=['No. of Null Values'])
```

# Data Wrangling Code


From columns like is_canceled,is_repeated_guest,reservation_status we replace
numbers into string.

Further, we drop some columns like agent,company,adr

```python
df['is_canceled'] = df['is_canceled'].replace([0,1],['Not Canceled','Canceled'])
```

```python
df['is_repeated_guest'] = df['is_repeated_guest'].replace([0,1],['No Previous Booking','Previous Booking'])
```

```python
df['reservation_status'] = df['reservation_status'].str.replace('No-Show','Canceled')
```

```python
df = df.drop(columns = {'agent','company','adr'})
```

# Stacked Bar Plot

```python
ax = sns.countplot(x='reservation_status',hue='hotel',data=df)
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.annotate(f'{height}', (x + width / 2, y + height / 2), ha='right', va='center')
```

In this plot,we conclude that resort hotel is more busy than city hotel.

# In this Plot, we plot the box plot by using seaborn library.



```python
sns.catplot(x='reservation_status',y='lead_time',hue='hotel',data=df,kind='box')
```

From this Plot, we analyse that if the difference between the number of days is more between the booking date and arrival date,then the chance of cancellation of booking is very high.

# Bar Plot

```python
H_C_R= df.groupby(['hotel','reservation_status'])['required_car_parking_spaces'].sum().reset_index()
```

```python
sns.barplot(x='reservation_status',y='required_car_parking_spaces',hue='hotel',data = H_C_R,palette=['blue','orange'])
```

From this plot,we analyse that car parking space is neccessary for both type of hotels to avoid cancellations of bookings..

# Stacked Bar Plot

```python
aggregated_data = df.groupby(['arrival_date_year','hotel','arrival_date_month','reservation_status']).size().unstack()
aggregated_data.plot(kind='bar', stacked=False, figsize=(12, 6))

plt.xlabel('Year, Month')
plt.ylabel('Count')
plt.title('Reservation Status by Year and Month')
plt.legend(title='Reservation Status')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

From This plot,we analyse check outs and cancellations for each month of each hotel.

By this, we say that those months which are less bookings as compared to others.
So we can offers to customers for those months to maximise bookings..

# Pie Plot

```python
dff = df[~(df['market_segment']=='Aviation')]
dfff = dff[~(dff['market_segment']=='Undefined')]
```

```python
pi_plot = dfff['market_segment'].value_counts()
```

```python
plt.figure(figsize = (10,5))

plt.pie(pi_plot,labels = pi_plot.index,autopct='%3.2f%%', startangle=140)
```

From this plot,we analyse that mostly customers preferred online platform for booking.

# Bar Plot

```python
s_d_f = df.groupby(['hotel','reservation_status'])['days_in_waiting_list'].mean()
```

```python
s_d_f
```

# Stacked Bar Plot

```python
a_g_t = df.groupby(['adults','babies','children','hotel','reservation_status']).size().unstack()
a_g_t.plot(kind='bar',stacked = False,figsize=(12, 6))
```

From this plot,we say that to avoid cancellation of bookings of each hotel,babies and children are allowed so that cancellation is not done.

# Stacked Bar Plot

```python
s = df.groupby(['hotel','deposit_type','reservation_status']).size().unstack()
```

```python
ax = s.plot(kind='bar',stacked = False)
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.annotate(f'{height}', (x + width / 2, y + height / 2), ha='center', va='center')
```

In this Plot,No Deposit process affect the cancellation of booking.

To avoid this type of cancellation,at the time of booking some amount of money deposit by customers.
